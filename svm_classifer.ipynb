```python
from sklearn import svm
import numpy
import types
import sys
import matplotlib.pyplot as plt
```

```python
## convert from list to numpy array
def convert_to_array(x):
    if isinstance(x, types.ListType):
        return numpy.array(x)
    else:
        print 'Not a list type'
```

```python
## open up a csv file and convert it to a numpy array
def read_and_clean(fname):
    data_array = []
    try:
        fp = open(fname, 'r')
    except:
        print 'File not found'
        sys.exit(-1)
 
    try:
        data = fp.readlines()
        for d in data:
            data_array.append(
                convert_to_array([float(i) for i in d.split(",")]))
        return data_array
    except:
        print 'Error in converting data'
        sys.exit(-1)
```
